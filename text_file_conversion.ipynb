```python
import matplotlib.pyplot as plt

x = [200,400,600,800,1000]

y1 = [12,20,34,45,55]
y2 = [15,25,36,50,60]
y3 = [18,30,40,55,70]
y4 = [20,35,45,60,80]
y5 = [25,40,53,75,100]


plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')
plt.plot(x, y3, label='Line 3', marker='^')
plt.plot(x, y4, label='Line 4', marker='d')
plt.plot(x, y5, label='Line 5', marker='x')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('performace comparison')


plt.legend()


plt.grid(True)
plt.show()
```
